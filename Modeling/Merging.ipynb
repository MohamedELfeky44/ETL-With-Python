```python
import pandas as pd 
```

```python
products = pd.read_csv("../Staging_2/products.csv")
products
```

```python
orders = pd.read_csv("../Staging_2/orders.csv")
orders
```

```python
order_items = pd.read_csv("../Staging_2/order_items.csv")
order_items
```

```python
merged_df = pd.merge(order_items, products, on='product_id', how='inner')
merged_df
```

```python
merged_df = pd.merge(merged_df, orders, on='order_id', how='inner')
merged_df
```

```python
merged_df.columns
```

```python
merged_df.columns = merged_df.columns.str.replace('_x', '')
merged_df.columns
```

```python
columns_to_drop = [col for col in merged_df.columns if col.endswith('_y')]
merged_df.drop(columns=columns_to_drop, inplace=True)
merged_df.columns
```

```python
import os

# Specify the folder path
folder_path = "../Information Mart"

# Check if the folder doesn't already exist
if not os.path.exists(folder_path):
    # Create the folder
    os.makedirs(folder_path)
    print(f"Folder '{folder_path}' created successfully.")
else:
    print(f"Folder '{folder_path}' already exists.")
```

```python
merged_df.to_csv("../Information Mart/orders_products_details.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame (assuming you have your data stored in a DataFrame called df)
# Replace this with your actual DataFrame
# df = ...

# Bar Chart of Order Status Counts
plt.figure(figsize=(8, 6))
sns.countplot(data=merged_df, x='order_status')
plt.title('Order Status Counts')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


```

```python
plt.figure(figsize=(3, 3))
sns.scatterplot(data=merged_df, x='list_price', y='quantity', hue='product_id', palette='viridis', s=100)
plt.title('Relationship between List Price and Quantity for Different Products')
plt.xlabel('List Price')
plt.ylabel('Quantity')
plt.legend(title='Product ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
order_status_counts = merged_df['order_status'].value_counts()

# Create a pie chart
plt.figure(figsize=(4, 4))
plt.pie(order_status_counts, labels=order_status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Order Statuses')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
