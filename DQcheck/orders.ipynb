```python
import pandas as pd 
file_path = '../Landing/orders.csv'
df = pd.read_csv(file_path)
df
```

```python
#check for nulls
df.isnull().sum()
```

```python
df.fillna('UNKNOWN',inplace =True)
df.isnull().sum()
```

```python
df.fillna('UNKWOWN',inplace = True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#Handel Date Range to be between 2000 and 2024
df['order_date'] = pd.to_datetime(df['order_date'],errors='coerce')
df['order_year'] = df['order_date'].dt.year
df
```

```python
df = df[(df['order_year'] >= 2000.0) & (df['order_year'] <= 2024.0) ]
df
```

### since there are no duplicates and we handeled nulls we can write the file

```python
import os

# Define the file path
output_directory = "../staging_1"
output_file_path = os.path.join(output_directory, "orders.csv")

# Write the DataFrame to CSV
df.to_csv(output_file_path, index=False)
```
