```python
import pandas as pd 
file_path = '../Landing/order_items.csv'
df = pd.read_csv(file_path)
df
```

```python
#check for nulls
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

### For order item ⇒ price within normal range (200:10000) + item_id is INT/float  and available in products


```python
df.info()
```

```python
df = df[ (df['list_price'] >= 200) | (df['list_price'] <= 10000)]
df
```

```python
file_path2 = '../Landing/products.csv'
df2 = pd.read_csv(file_path)
df2
```

```python
df = df[df['item_id'].isin(df2['item_id'])]
df
```

### since there are no nulls or duplicates we can write the file

```python
import os

# Define the file path
output_directory = "../staging_1"
output_file_path = os.path.join(output_directory, "order_items.csv")

# Write the DataFrame to CSV
df.to_csv(output_file_path, index=False)
```
