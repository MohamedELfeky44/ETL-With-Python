```python
import pandas as pd

# Use backslashes in the file path
csv_file_path = "../source/orders.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

df
```

```python
table_name = 'orders'
df.to_sql(table_name, con=engine, if_exists='replace', index=False)
```

```python

# Use backslashes in the file path
csv_file_path = "../source/order_items.csv"

# Read the CSV file into a DataFrame
df2 = pd.read_csv(csv_file_path)

df2
```

```python
from sqlalchemy import types

dtype_mapping = {
    'order_id': types.NUMERIC(precision=20, scale=4),
    'items_id': types.NUMERIC(precision=20, scale=4),
    'product_id': types.NUMERIC(precision=20, scale=4),
    'quantity': types.NUMERIC(precision=20, scale=4),
    'list_price': types.FLOAT,
    'discount': types.FLOAT
}

table_name = "order_items"
df2.to_sql(table_name, con=engine, index=False, if_exists='replace', dtype=dtype_mapping)
```

```python
sql_query = 'SELECT * FROM orders'

# Execute the SQL query and fetch the results into a DataFrame
df_from_db = pd.read_sql(sql_query, connection)
df_from_db
```

```python
import os
import shutil
source_directory = "../source"
target_directory = "../bikestore_dim"
os.makedirs(target_directory)
```

```python
files = os.listdir(source_directory)
exclude_files = ['order.csv', 'order_items.csv']
exclude_directories = ['.ipynb_checkpoints']  
for file in files:
    if file not in exclude_files and file not in exclude_directories:
        source_path = os.path.join(source_directory, file)
        target_path = os.path.join(target_directory, file)
        shutil.copy(source_path, target_path)
```
